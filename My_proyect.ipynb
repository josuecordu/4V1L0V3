```python
# Instalar spaCy y transformers
!pip install spacy
!pip install transformers

# Descargar los modelos de spaCy para inglés y español
!python -m spacy download en_core_web_sm
!python -m spacy download es_core_news_sm
import spacy
from transformers import pipeline

# Cargar los modelos de spaCy
nlp_en = spacy.load("en_core_web_sm")
nlp_es = spacy.load("es_core_news_sm")

# Cargar el pipeline de transformers para resumen
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Texto de ejemplo en inglés
texto_en = """
Google Colab is a great tool for data science and machine learning. It allows you to run Python code in the cloud without the need for a local setup.
"""

# Texto de ejemplo en español
texto_es = """
Google Colab es una herramienta excelente para la ciencia de datos y el aprendizaje automático. Permite ejecutar código Python en la nube sin necesidad de una configuración local.
"""

# Función para procesar texto: extraer entidades y generar resumen
def procesar_texto(texto, idioma='es'):
    # Elegir el modelo spaCy según el idioma
    doc = nlp_es(texto) if idioma == 'es' else nlp_en(texto)

    # Extraer las entidades del texto
    entidades = [(ent.text, ent.label_) for ent in doc.ents]

    # Generar el resumen con transformers
    resumen = summarizer(texto, max_length=150, min_length=50, do_sample=False)

    return resumen[0]['summary_text'], entidades

# Procesar el texto en inglés
resumen_en, entidades_en = procesar_texto(texto_en, 'en')

# Procesar el texto en español
resumen_es, entidades_es = procesar_texto(texto_es, 'es')

# Mostrar resultados
print("Resumen en inglés:", resumen_en)
print("Entidades en inglés:", entidades_en)

print("\nResumen en español:", resumen_es)
print("Entidades en español:", entidades_es)
```
